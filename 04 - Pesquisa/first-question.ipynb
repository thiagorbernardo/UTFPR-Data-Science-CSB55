```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import os
import statsmodels.stats.api as sms
from statsmodels.formula.api import ols
pd.set_option('display.max_columns', None)
```

```python
df_emissions = pd.read_csv('../data/Gas Emissions/total-ghg-emissions.csv',comment='#', low_memory=False)
df_emissions.drop(df_emissions[
    (df_emissions['Entity'] != 'South America') &
    (df_emissions['Entity'] != 'North America') &
    (df_emissions['Entity'] != 'Europe') &
    (df_emissions['Entity'] != 'Africa') &
    (df_emissions['Entity'] != 'Oceania') &
    (df_emissions['Entity'] != 'Asia')
].index, inplace=True)
df_emissions.drop('Code', axis=1, inplace=True)
df_emissions.columns = ['REGION', 'DATE', 'EMISSIONS']
```

```python
df_emissions.REGION.unique()
```

```python
df_emissions.head(3)
```

```python
df_america = df_emissions[
    (df_emissions["REGION"] == 'North America') |
    (df_emissions["REGION"] == 'South America')
].groupby('DATE').sum().reset_index()

df_america["REGION"] = "Americas"
df_america.head(3)
```

```python
df_emissions = df_emissions.append(df_america)
df_emissions.REGION.unique()
```

```python
df_iqar = pd.read_csv('../data/air-quality/results/iqar-continents-class-after.csv',comment='#', low_memory=False)
```

```python
df_iqar
```

```python
df = pd.merge(df_iqar, df_emissions,  how='inner', on=['REGION','DATE'])
df.head(10)
```

### Removendo dados de 2020+

```python
df = df.drop(df[df.isna().any(axis=1)].index)
df.tail(8)
```

```python
df.CLASSIFICATION.unique()
```

## Modelando com base nas emiss√µes

```python
from sklearn.preprocessing import MinMaxScaler

df_sklearn = df.copy()
  
df_sklearn['CLASSIFICATION'] = MinMaxScaler().fit_transform(np.array(df_sklearn['CLASSIFICATION']).reshape(-1,1))
df_sklearn['EMISSIONS'] = MinMaxScaler().fit_transform(np.array(df_sklearn['EMISSIONS']).reshape(-1,1))
  
display(df_sklearn)
```

```python
import statsmodels.formula.api as smf
model = smf.ols("CLASSIFICATION ~ EMISSIONS", data=df_sklearn)
response = model.fit()
response.summary()
```
